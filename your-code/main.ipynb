## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# your code here
import scipy.stats as stats

# Given parameters
low = 5  # Lower bound
high = 10  # Upper bound
size = 1000  # Number of random samples to generate

# Generate uniform random numbers
uniform_random_numbers = stats.uniform.rvs(loc=low, scale=high-low, size=size)

# Show first 10 numbers as a sample
print(uniform_random_numbers[:10])
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
def generate_uniform_random_numbers(bottom, ceiling, count):
    return np.random.uniform(bottom, ceiling, count)
```

```python
# your code here
random_numbers_1 = generate_uniform_random_numbers(10, 15, 100)
random_numbers_2 = generate_uniform_random_numbers(10, 60, 1000)
```

```python
# your code here
plt.figure(figsize=(12, 6))

# Plot for first set (bottom=10, ceiling=15, count=100)
plt.subplot(1, 2, 1)
plt.hist(random_numbers_1, bins=10, color='skyblue', edgecolor='black')
plt.title('Uniform Distribution (10 to 15, 100 numbers)')
plt.xlabel('Value')
plt.ylabel('Count')

# Plot for second set (bottom=10, ceiling=60, count=1000)
plt.subplot(1, 2, 2)
plt.hist(random_numbers_2, bins=10, color='skyblue', edgecolor='black')
plt.title('Uniform Distribution (10 to 60, 1000 numbers)')
plt.xlabel('Value')
plt.ylabel('Count')

# Display the plots
plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# your answer below
#First Distribution: The histogram have fewer data points per bin, so it appear more scattered.
#Second Distribution: The histogram show more evenly distributed values, resulting in a smoother and more stable distribution.
```

```python
# your code here
def plot_qq(data, title="Q-Q Plot"):
    """
    Function to plot a Q-Q plot for a given dataset to assess normality.

    Parameters:
    - data: The data variable (Series or list/array).
    - title: The title for the plot.
    """
    plt.figure(figsize=(6, 6))
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(title)
    plt.grid(True)
    plt.show()

# Example usage with your data:
plot_qq(vehicles['Fuel Barrels/Year'], title="Q-Q Plot for Fuel Barrels/Year")
```

```python
plot_qq(vehicles['CO2 Emission Grams/Mile'], title="Q-Q Plot for CO2 Emission Grams/Mile")
```

```python
plot_qq(vehicles['Combined MPG'], title="Q-Q Plot for Combined MPG")
```

```python
def shapiro_wilk_test(data):
    """
    Perform Shapiro-Wilk test for normality.

    Parameters:
    - data: The dataset (Series or list/array).

    Returns:
    - The test statistic and the p-value.
    """
    stat, p_value = stats.shapiro(data)
    print(f"Shapiro-Wilk Test Statistic: {stat}")
    print(f"p-value: {p_value}")

    if p_value > 0.05:
        print("The data looks normal (fail to reject H0).")
    else:
        print("The data does not look normal (reject H0).")

# Example usage with your data:
shapiro_wilk_test(vehicles['Fuel Barrels/Year'])
shapiro_wilk_test(vehicles['CO2 Emission Grams/Mile'])
shapiro_wilk_test(vehicles['Combined MPG'])
```

None of them are normally ditributed. 

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here

def generate_exponential_distribution(mean, size):
    """
    Generate a list of exponentially distributed numbers.
    
    Parameters:
    - mean: The mean of the exponential distribution.
    - size: The size of the sequence to generate.
    
    Returns:
    - A numpy array of exponentially distributed random numbers.
    """
    return np.random.exponential(mean, size)

# Generate two sequences with different sizes
sequence_10 = generate_exponential_distribution(10, 10)
sequence_100 = generate_exponential_distribution(10, 100)

# Plot the histograms with 100 bins
plt.figure(figsize=(12, 6))

# First subplot: Histogram for size 10
plt.subplot(1, 2, 1)
plt.hist(sequence_10, bins=100, color='skyblue', edgecolor='black')
plt.title("Exponential Distribution (size=10)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Second subplot: Histogram for size 100
plt.subplot(1, 2, 2)
plt.hist(sequence_100, bins=100, color='lightgreen', edgecolor='black')
plt.title("Exponential Distribution (size=100)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#How are the two distributions different?

#Smaller sample (10): Less smooth, more statistical noise, and a less accurate representation of the exponential distribution.
#Larger sample (100): Smoother, more accurate, and a better representation of the expected distribution.
```

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
lambda_ = 1 / 10  # Rate parameter (1/mean)
x = 15  # Time in minutes for which we want to calculate the probability

# CDF formula for exponential distribution
probability = 1 - np.exp(-lambda_ * x)
```

```python
# your answer here
# Hint: This is same as saying P(x<15)
print(f"The probability that a customer will spend less than {x} minutes in the bank is: {round(probability * 100, 4)}%")
```

What is the probability that the customer will spend more than 15 minutes

```python
# your code here
# Probability that the customer will spend more than 15 minutes
probability_more_than_15 = 1 - probability
print(f"The probability that a customer will spend more than {x} minutes in the bank is: {round(probability_more_than_15 * 100, 4)}%")
```
